```python
import csv
lines = []
with open('titanic_train.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for line in csv_reader:
        lines.append(line)
data = lines[1:]
passengers = []
headers = lines[0]

# Store the values in a list
for d in data:
    p = {}
    for i in range(0, len(headers)):
        key = headers[i]
        value = d[i]
        p[key] = value
        passengers.append(p)

# Extract fields into their respective lists
survived = [p['Survived'] for p in passengers]
pclass = [p['Pclass'] for p in passengers]
age = [float(p['Age']) for p in passengers if p['Age'] != '']
gender_survived = [p['Sex'] for p in passengers if int(p['Survived']) != 1]
```

```python
# Proportion of passengers that survived the incident
import matplotlib.pyplot as plt
nums = [0,0]
for i in range(len(survived)):
    if int(survived[i]) == 0:
        nums[0] += 1
    else:
        nums[1] += 1
plt.pie(nums, labels=['0', '1'], autopct='%1.1f%%', colors=['lightblue', 'yellow'])
plt.title('Survived', fontdict={'fontsize':20})
plt.show()
```

```python
# Compare the gender of passengers
import matplotlib.pyplot as plt
gender = ['female','male']
nums = [0,0]
for i in range(len(gender_survived)):
    if gender_survived[i] == 'female':
        nums[0] += 1
    else:
        nums[1] += 1
plt.bar(gender, nums)
plt.title('surviving passengers count by gender')
plt.show()
```
